```python
def pangrams(s):
    if len(s)<26:
        return "not pangram"
    
    s = s.lower()
    dic = {chr(i):0 for i in range(97, 123)}
    
    
    for i in s:
        if i in dic.keys():
            dic[i] = dic[i]+1
        
    if 0 in dic.values():
        return "not pangram"
    else:
        return "pangram"
```
