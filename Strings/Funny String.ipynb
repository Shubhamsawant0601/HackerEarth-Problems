```python

def funnyString(s):
    s = [ord(i) for i in s]
    r = s[::-1]
    
    for i in range(len(s)-1):
        if abs(s[i+1]-s[i]) != abs(r[i+1]-r[i]):
            return "Not Funny"
            
    return "Funny"
```

```python
s="acxz"
funnyString(s) 
```
