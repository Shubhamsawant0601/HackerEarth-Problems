```python
def libraryFine(d1, m1, y1, d2, m2, y2):
    fine = 0
    
    if y2<y1:
        fine = 10000
        
    elif y1<y2:
        fine = 0
        
    else:
        if m2<m1:
            fine = 500*(m1-m2)
        elif m1<m2:
            fine = 0
        else:
            if d2<d1:
                fine = 15*(d1-d2)
            elif d1<d2:
                fine = 0
            else:
                fine = 0
                
    return fine
```

```python
libraryFine(2, 7, 2015, 1, 2, 2014)
```
