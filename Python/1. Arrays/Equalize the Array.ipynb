```python
def equalizeArray(arr):
    # Counter Dictionary
    dic = {}
    for i in arr:
        if i in dic.keys():
            dic[i] += 1
        else:
            dic[i] = 1
            
    # Sorted Dictionary for getting maximum count of element
    dic = dict(sorted(dic.items(), key = lambda x:x[1], reverse=True))
    
    common_values = list(dic.values())
     
    # Elements to deleted = Total Element - Count of Most common element
    deletion = len(arr) - common_values[0]
    
    return (deletion)
```
