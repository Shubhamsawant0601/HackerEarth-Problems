```python
def circularArrayRotation(a, k, queries):
    li = []
    k = k%len(a)
    
    for i in range(k):
        a = rotate(a)
        
    for i in queries:
        li.append(a[i])
        
    return li
    
def rotate(arr):
    a = arr[len(arr)-1]
    
    arr.pop()
    
    arr.insert(0, a)
    
    return arr
```

```python
a = [1,2,3,4,5,6,7]
k = 5
queries = [1,2,3]


circularArrayRotation(a, k, queries)
```
