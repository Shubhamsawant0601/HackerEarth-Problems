```python
def beautifulDays(i, j, k):
    count = 0
    for n in range(i,j+1):
        if isBeautiful(n, k):
            count+=1
    return count
    
    
def isBeautiful(n, k):
    reverse = str(n)[::-1]
    reverse = int(reverse)
    diff = abs(n - reverse)
    
    num = float(diff)/k
    
    if num%1==0 :
        return True
    else:
        return False
```

```python
i = 20
j = 23
k = 6
beautifulDays(i, j, k)
```
