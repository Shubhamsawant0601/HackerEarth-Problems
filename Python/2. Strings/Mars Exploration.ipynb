```python
def marsExploration(s):
    exp = expacted(s)
    count = 0
    for i, j in zip(s, exp):
        if i!=j:
            count+=1
    
    return count
    
    
def expacted(s):        
    li = []
        
    for i in range(1, len(s)+1):
        
        if i%3==0 or i%3==1 :
            li.append("S")
            
        if i%3==2:
            li.append("O")
            
            
    return (li)
```

```python
s = "SOSOSSOSSOSSOSSOS"
marsExploration(s)
```
